```python
import pandas as pd
from time import sleep
from random import randint
import spotipy 
from spotipy.oauth2 import SpotifyClientCredentials
import warnings
warnings.filterwarnings('ignore')
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from IPython.display import Image
Image("Spotify_Logo_RGB_Black.png")


secrets_file = open("secrets.txt","r")

string = secrets_file.read()
secrets_dict={}

for line in string.split('\n'):
    secrets_dict[line.split('=')[0]]=line.split('=')[1]
    
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= secrets_dict['cid'],
                                                           client_secret=secrets_dict['c_secret']))


print("\n\n**************** WELCOME TO THE SPOTIFY RIPPER ****************\n\n\nGet your playlist ID at https://open.spotify.com/search \n\n\nThen copy the code situated after #https://open.spotify.com/playlist/\n\n")

t = randint(1,21)

playlist_name = input('Enter the playlist name : ')
playlist_user_input = input('\nCopy/Paste your playlist ID here :')
playlist = sp.user_playlist_tracks("spotify", playlist_user_input)
play_check = sp.user_playlist_tracks("spotify", playlist_user_input)

def user_input():
    if playlist_user_input == playlist_user_input:
        return print('\n\nPlaylist Name :',playlist_name,'\n\n\nPlaylist ID (in use) :', playlist_user_input,'\n\n\n*******************************************************************************************************************************\n\n>>>>>  Test your playlist avalaible info several times before running (100 first songs) (x3)\n' )
user_input()

playlist_count = []
```

```python
##################################################################################################################################
#################################################### RUN TEST ####################################################################
##################################################################################################################################
from IPython.display import Image
Image("api_100.png")

playlist_count = []

while play_check['next']!=None:
    play_check = sp.next(play_check)
    for i in range(len(play_check["items"])):
            count = play_check["items"][i]["track"]["name"]
            playlist_count.append(count)


t = randint(1,33)
print('\n\\\TESTING FIRST THIRD **************************************************************************************** Testing 1/3 ///')
sleep(1)
print('*****************  Track Test:', playlist["items"][t]["track"]["name"],' ************************************')
sleep(1)
print('*******************************  Artist Test:', playlist["items"][t]["track"]['album']["artists"][0]["name"],' **********************************************************')
sleep(1)
print('*************************************************  Uri Test:', playlist["items"][t]["track"]["uri"],' ***************************')
sleep(1)
print('*******************************************************************************************************************************')
sleep(1)
print('*******************************  Audio Features Test:', (sp.audio_features(playlist["items"][t]["track"]["uri"])))
sleep(1)
print('******************  TEST 1 COMPLETE  ******************************************************************************************')
sleep(1)

t2 = randint(34,67)
print('\n\n\\\TESTING SECOND THIRD *************************************************************************************** Testing 2/3 ///')
sleep(1)
print('*****************  Track Test:', playlist["items"][t2]["track"]["name"],' ****************************************')
sleep(1)
print('*******************************  Artist Test:', playlist["items"][t2]["track"]['album']["artists"][0]["name"],' ***********************************************')
sleep(1)
print('**************************************************  Uri Test:', playlist["items"][t2]["track"]["uri"],' ***************************')
sleep(1)
print('*******************************************************************************************************************************')
sleep(1)
print('*******************************  Audio Features Test:', (sp.audio_features(playlist["items"][t2]["track"]["uri"])))
sleep(1)
print('*****************  TEST 2 COMPLETE  *******************************************************************************************')
sleep(1)

t3 = randint(68,100)
print('\n\n\\\TESTING LAST THIRD ***************************************************************************************** Testing 3/3 ///')
sleep(1)
print('*****************  Track Test:', playlist["items"][t3]["track"]["name"],' *************************************')
sleep(1)
print('*******************************  Artist Test:', playlist["items"][t3]["track"]['album']["artists"][0]["name"],' ***************************************************')
sleep(1)
print('**************************************************  Uri Test:', playlist["items"][t3]["track"]["uri"],' ***************************')
sleep(1)
print('*******************************************************************************************************************************')
sleep(1)
print('*******************************  Audio Features Test:', (sp.audio_features(playlist["items"][t3]["track"]["uri"])))
sleep(1)
print('****************  FINAL TEST COMPLETE  ***************************************************************************************')
sleep(4)
print('\n\n\n\n                                         You have',len(playlist_count),'songs in your playlist.')
sleep(2)
print('\n\n\n                                          |                               |\n                                          |                               |\n                                          |                               |\n                                          |                               |\n                                          |     RUN THE API EXTRACTOR     |\n                                          |                               |\n                                          |                               |\n                                          V                               V')

```

```python
###############################################################################################################################
###############################################################################################################################
################################################# API EXTRACTOR ###############################################################
###############################################################################################################################
###############################################################################################################################
from IPython.display import Image
Image("ripper_banner.png")


playlist_audio_features = []
playlist_tracks = []
playlist_artist = []

print('You have',len(playlist_count),'requests pending.')
if len(playlist_count) > 3000:
    print('\n!!!!!!!!!!!!WARNING!!!!!!!!!!!!! Too large playlist may have issues !!!. Your playlist')
    api_extractor_input = input('\nRequest Audio Features (Y/N)  ? :')
else:
    api_extractor_input = input('\nRequest Audio Features (Y/N)  ? :')
sleep(1)


while playlist['next']!=None:
    if api_extractor_input == 'Y' or api_extractor_input == 'y':
        print('Requesting...')
        chunk_of_playlist = []
        playlist = sp.next(playlist)
        for i in range(len(playlist["items"])):
            tracks = playlist["items"][i]["track"]["name"]
            playlist_tracks.append(tracks)
        for i in range(len(playlist["items"])):
            artist = playlist["items"][i]["track"]['album']["artists"][0]["name"]
            playlist_artist.append(artist)
        for i in range(len(playlist["items"])):    
            song_uri = playlist["items"][i]["track"]["uri"]
            chunk_of_playlist.append(song_uri)
        playlist_audio_features.append(chunk_of_playlist)
        sleep(2)
    elif api_extractor_input == 'N' and api_extractor_input == 'n':
        break
    else:
        alternate_ID = input('Enter alternate playlist ID and run again:')
        playlist = sp.user_playlist_tracks("spotify", alternate_ID)  



print('\nSongs collected :', len(playlist_tracks))
sleep(2)
if len(playlist_tracks) == len(playlist_count):
    print('\n100% of resquests were successful.')
    sleep(2)
    print('\n\n\n>>>>>>> Run the integrity test to check if the API extractor ripped your playlist correctly.')
elif len(playlist_tracks) != len(playlist_count):
    print('\nOnly',(len(playlist_tracks)/len(playlist_count)*100),'% of the whole playlist had been collected.')
    print('\You can still proceed to extraction or restart the whole process.')
else:
    print('\n\n\n>>>>>>> Run the integrity test to check if the API extractor ripped your playlist correctly.')
```

```python
###############################################################################################################################
################################################# INTEGRITY TEST ##############################################################
###############################################################################################################################




from IPython.display import Image
Image("integrity.png")



track_items = len(playlist_tracks)
artist_items = len(playlist_artist)
uri_sublists = len(playlist_audio_features)
uri_sublists_items = len(playlist_audio_features[0])
last_list_items = len(chunk_of_playlist)
uri_total = ((uri_sublists-1)*uri_sublists_items) + last_list_items

print('\n********************************************* API EXTRACTOR INTEGRITY TEST ****************************************************')
sleep(1)
print('*******************************************************************************************************************************')
sleep(1)
print('*****************  Track Items:', track_items,' ****************************************************************************************')
sleep(1)
print('******************************  Artist Items:', artist_items,' ***************************************************************************')
sleep(1)
print('********************************************  Uri Sub-Lists:', uri_sublists,' *************************************************************')
sleep(1)
print('************************************************************ Uri Sub-List Items:', uri_sublists_items,' *****************************************')
sleep(1)
print('****************************************************************************  Uri Total:', uri_total,'   *****************************')
sleep(1)
print('*******************************************************************************************************************************')
if track_items == artist_items & artist_items==uri_total & uri_total==track_items :
    print('*** Integrity 100%: Please proceed to the next step.')
    sleep(3)
    print('\n\n\n\n\n                                          | Extract the data contained in URI |\n                                          V                                   V')
else:
     print('Too bad!!! :( Integrity compromised. Run the ripper again. XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX///RESTART KERNEL AND RUN FROM THE 1ST STEP')
```

```python
###############################################################################################################################
###############################################################################################################################
################################################# URI EXTRACTOR ###############################################################
###############################################################################################################################
###############################################################################################################################

from IPython.display import Image
Image("ripper_banner.png")

uri_extracted = []

print('\n*****************************************************************************************************************************')
print('\nYou have',uri_total,'URI to extract.')

if uri_total > 3000:
    print('!!!!!!!!!!!!WARNING!!!!!!!!!!!!! Large playlist will take a significant amount of time !!!')
    uri_extractor_input = input('\nExtract Audio Features (Y/N)  ? :')
else:
    uri_extractor_input = input('\n>>>>>>>>>> U.R.I EXTRACTOR >>>>>>> Extract Audio Features (Y/N)  ? :')
sleep(1)

 
for sublist in playlist_audio_features:
    if uri_extractor_input == 'Y' or uri_extractor_input == 'y':
        uri_extracted.append(sp.audio_features(sublist))
        print('Extracting...')
    elif uri_extractor_input == 'N' or uri_extractor_input == 'n':
        print('Your content has not been extracted.')
    else:
        uri_extractor_input = input('\nType Error : Extract Audio Features (Y/N)  ? :')
    sleep(2)
    
print('\n\nDone.',uri_total,'Audio Features extracted.')
sleep(1)
print('...')
sleep(1)
print('List of Audio Features has been flattened.')
sleep(1)
print('...')
sleep(1)
print('...')
sleep(1)
print('...')
sleep(2)
print('Your dataframe is stored as <data>.')
sleep(2)


def flatten(input_list):
    return[item for sublist in input_list for item in sublist]

final_audio_features = flatten(uri_extracted)

playlist_audio_features_df = pd.DataFrame(final_audio_features)
playlist_tracks_df = pd.DataFrame(playlist_tracks)
playlist_artist_df = pd.DataFrame(playlist_artist)

data = pd.concat((playlist_tracks_df, playlist_artist_df, playlist_audio_features_df),axis=1)
data = data.drop(['type','id','uri','track_href','analysis_url'], axis=1)
data.columns = [                 'title',                  'artist',     'danceability',
                 'energy',              'key',         'loudness',
                   'mode',      'speechiness',     'acousticness',
       'instrumentalness',         'liveness',          'valence',
                  'tempo',             
            'duration_ms',   'time_signature']


rows = data.shape[0]
columns = data.shape[1]
print("\nThe new dataset contains {0} rows and {1} columns.".format(rows, columns))
sleep(2)
print('The dataset has', data['artist'].nunique(),'artists referenced.')
sleep(2)
print('It has an average of',((data.isnull().sum()/data.shape[0])*100).mean(),'% of missing values.')
sleep(2)    
if len(data['title']) > data['title'].nunique() or data.isnull() != 0:
    print('\nIt may have',len(data['title']) - data['title'].nunique(),'songs that appears twice or more.')
    sleep(2)
    data_info_input = input('Perform cleaning operations ? (Y/N):')
    sleep(2)
    if data_info_input == 'Y' or data_info_input == 'y':
        data = data.drop_duplicates()
        sleep(1)
        data = data.dropna()
        print('Cleaning dataset..')
        sleep(3)
        print('...')
        sleep(2)
        print('The cleaned dataset has now', len(data['title']),'songs and',data['artist'].nunique(),'artists referenced.')
        sleep(2)
        print('It has now an average of',((data.isnull().sum()/data.shape[0])*100).mean(),'% of missing values.')
        sleep(3)
        print('\n*** Ready to export. Thanks for using the Audio Features Ripper.')
    elif data_info_input == 'N' or data_info_input == 'n':
        print('\nWarning !!! : Your dataset will contain',data['title'] - data['title'].nunique(),'potential duplicates with an average of'((data.isnull().sum()/data.shape[0])*100).mean(),'NaN values.')
        sleep(2)
        print('\n*** Ready to export anyway. Thanks for using the Audio Features Ripper.')
        sleep(2)
    else:
        print('Please type your answer again (Y/N) : ')
        data_info_input
else:
    print('\n*** Ready to export. Thanks for using the Audio Features Ripper.')
    
data.head(10)


print('\n\nYour playlist name is :',playlist_name,'\n\n\n****************************************** Name your CSV file and export ****************************************************')
```
